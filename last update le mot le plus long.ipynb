```python
import random
import time

# Définition des sacs de lettres
voyelles = ['a', 'e', 'i', 'o', 'u', 'y']
consonnes = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'z', 'ç']

# Définition de la fonction pour piocher des lettres dans le sac
def piocher_lettre(sac):
    return random.choice(sac)

# Définition de la fonction pour jouer une manche
def jouer_manche():
    # Initialisation du jeu
    lettres = []
    for i in range(10):
        if i % 2 == 0:
            lettre = piocher_lettre(voyelles)
        else:
            lettre = piocher_lettre(consonnes)
        lettres.append(lettre)
    print("Lettres disponibles : ", lettres)

    # Définition de la durée limite de temps pour chaque manche (ici, 30 secondes)
    temps_limite = 30
    debut = time.time()  # temps de départ de la manche
    temps_ecoule = 0  # initialisation du temps écoulé à 0
    mot = None  # initialisation du mot à None
    while not mot and temps_ecoule < temps_limite:  # tant que le mot n'est pas saisi et le temps n'est pas écoulé
        temps_restant = int(temps_limite - temps_ecoule)
        print("Temps restant : ", temps_restant, "s")
        time.sleep(1)  # attendre 1 seconde
        mot = input("Entrez un mot de 2 à 10 lettres : ")
        temps_ecoule = time.time() - debut  # temps écoulé depuis le début de la manche
        if len(mot) < 2 or len(mot) > 10:
            print("Mot invalide !")
            mot = None  # réinitialisation du mot
    if temps_ecoule >= temps_limite:  # si le temps est écoulé
        print("Temps écoulé !")
        return 0  # retourne 0 points

    # Le mot est saisi avant la fin du temps imparti
    points = len(mot)
    if mot not in dictionnaire:
        points = 0
        print("Le mot", mot, "n'est pas dans le dictionnaire.")
    print("Le mot proposé est", mot, "et rapporte", points, "points.")
    return points

# Définition du dictionnaire
with open('Liste.txt', 'r') as f:
    dictionnaire = f.read().splitlines()

# Jouer une partie
score_joueur1 = 0
score_joueur2 = 0
for i in range(2):
    print("Manche", i+1)
    points_joueur1 = jouer_manche()
    points_joueur2 = jouer_manche()
    if points_joueur1 > points_joueur2:
        print("Le joueur 1 remporte la manche.")
        score_joueur1 += points_joueur1
    elif points_joueur2 > points_joueur1:
        print("Le joueur 2 remporte la manche.")
        score_joueur2 += points_joueur2
    else:
        print("Égalité.")
        score_joueur1 += points_joueur1
        score_joueur2 += points_joueur2

# Affichage du score final
print("Score final : Joueur 1 =", score_joueur1, ", Joueur 2 =", score_joueur2)
if score_joueur1 > score_joueur2:
    print("Le joueur 1 remporte la partie !")
elif score_joueur2 > score_joueur1:
    print("Le joueur 2 remporte la partie !")
else:
    print("Match nul !")
```
